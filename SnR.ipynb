```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from tqdm import tqdm
import mplfinance as mpf
import warnings
warnings.filterwarnings("ignore")
```

```python
## Dataframe manipulation
currentDir = os.path.dirname(os.path.abspath(__name__))
relativePath = 'data/ethData/ETHUSDT_12h.csv'
file_path= os.path.join(currentDir, relativePath)
df = pd.read_csv(file_path)
df = df.drop("Unnamed: 0", axis=1)
df = df.rename(columns={"datetime":"Date"})
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
```

```python
# mpf style
mc = mpf.make_marketcolors(up='#00ff00', down="#ff0000", inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='nightclouds', marketcolors=mc)
```

```python
df_for_plot = df[:200]
# Create candlestick chart
mpf.plot(df_for_plot, type='candle', style=s, title="ETH/USD 12H", figratio=(20,8), update_width_config=dict(candle_linewidth=0.8), volume=True)
```
